```python
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt
rng = np.random.default_rng() # create a random number generator object

N=10
T = 100
del_x = 1
x_max = 100
t = np.linspace(1,T,T+1)
loc = [0]
loc_all = []

#x = np.linspace(-x_max,x_max, del_x)

for n in range(N):
    for i in range(T):
        walk = rng.random()
        if round(walk) >= 0.5:
            loc.append(loc[i]+1)
        else:
            loc.append(loc[i]-1) 
    loc_all.append(loc)
    plt.plot(t,loc_all[n])
    loc = [0]

plt.xlabel("Time")
plt.ylabel("Location")
plt.title("1-D Random Walk")
plt.show()
```
