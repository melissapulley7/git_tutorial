```python
import numpy as np
import matplotlib.pyplot as plt

N = np.linspace(0,15,100)
alpha = 1/5
Total_time = 1000
h_time = 10
HA = Total_time/h_time

V = alpha*Total_time*N / (1+alpha*h_time*N)

plt.plot(N,V,label="V")#"$V =\frac{a} T N}{1+a * T_h *N}$")
plt.axhline(Total_time/h_time,linestyle="--", label="L2")
plt.axis([0,max(N),0,max(V)+h_time])
plt.xlabel("Number (or Density) of Prey")
plt.ylabel("Number of Prey Captured, Victims")
plt.title("Type II Functional Response Curve")
#plt.legend()
plt.savefig("typeII")
plt.show()
```
